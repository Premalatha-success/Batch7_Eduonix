#### Data Preparation

```python
df.info()
```

```python
df=df.drop(["PassengerId","Ticket","Name","Cabin"],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
age_median=df["Age"].median()
age_median
```

```python
df["Age"]=df["Age"].replace(np.nan,age_median)
```

```python
df.isnull().sum()
```

```python
emb_mode=df["Embarked"].mode()[0]
emb_mode
```

```python
df["Embarked"]=df["Embarked"].replace(np.nan,emb_mode)
```

```python
df.isnull().sum()
```

```python
sns.boxplot(x="Age",data=df)
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([.25,.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lowerage,upperage=remove_outlier(df["Age"])
df["Age"]=np.where(df["Age"]>upperage,upperage,df["Age"])
df["Age"]=np.where(df["Age"]<lowerage,lowerage,df["Age"])
```

```python
sns.boxplot(x="Age",data=df)
```
